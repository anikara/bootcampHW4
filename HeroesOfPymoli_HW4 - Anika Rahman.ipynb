```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load ='/Users/anikarahman/Desktop/Bootcamp_HW/HW4/purchase_data.csv'

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)



purchase_data.head()
```

## Player Count

* Display the total number of players


```python
totalplayers = len(purchase_data['SN'].unique())

pd.DataFrame([{'Total Players': totalplayers}])



```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
uniqueitems = len(purchase_data['Item ID'].unique())
avgprice = round(purchase_data['Price'].mean(),2)
num_purchases = purchase_data['Price'].count()
totrevenue = purchase_data['Price'].sum()
pd.DataFrame([{'Number of Unique Items': uniqueitems,
               'Average Price' : avgprice,
               'Number of Purchases' : num_purchases,
               'Total Revenue': totrevenue
              
              
              }]) 


```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gendertally = purchase_data.groupby([ 'SN', 'Gender']).count()
gendertally = gendertally.reset_index()
#a = gendertally['Gender'].value_counts()
#a = s['Gender'].count()


genders= pd.DataFrame(gendertally['Gender'].value_counts())
genders = genders.rename(columns={"Gender": "Total Count"})
genders['Percentage of Players'] = round(genders['Total Count']/totalplayers*100,2)
genders.head()

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
totalpurchasevalue = purchase_data.groupby('Gender')['Price'].sum()
avgpurchaseprice = round(purchase_data.groupby('Gender')['Price'].mean(),2)
purchasecount = purchase_data['Gender'].value_counts()
avgtotalperperson = round((purchase_data.groupby('Gender')['Price'].sum())/(genders['Total Count']),2) #####



puranalysis = pd.DataFrame(totalpurchasevalue) 
puranalysis = puranalysis.rename(columns={"Price": "Purchase Count"})
puranalysis['Average Purchase Price'] = avgpurchaseprice
puranalysis['Total Purchase Value'] = purchasecount
puranalysis['Avg Total Purchase per Person'] = avgtotalperperson
puranalysis['Avg Total Purchase per Person'] = puranalysis['Avg Total Purchase per Person'].map("${:.2f}".format)
puranalysis['Average Purchase Price'] = puranalysis['Average Purchase Price'].map("${:.2f}".format)
puranalysis['Total Purchase Value'] = puranalysis['Total Purchase Value'].map("${:.2f}".format)
puranalysis['Purchase Count'] = puranalysis['Purchase Count'].map("${:,.2f}".format)

puranalysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 10, 15,20,25,30,35,40,100]
grouplabels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

binvalue = pd.cut(purchase_data['Age'], bins, right = False, labels=grouplabels)
agedata = purchase_data
agedata['Total Count'] = binvalue


agedata2 = agedata.groupby(['SN', 'Total Count'])

counts = pd.DataFrame(agedata2['Total Count'].count())
counts['Total Count'] = 1
counts = counts.rename(columns = {"Total Count" : "zz"})
counts = counts.reset_index()
totalbincounts = pd.DataFrame(counts['Total Count'].value_counts())
totalbincounts['Percentage of Players'] = round(totalbincounts['Total Count']/totalplayers*100,2)
totalbincounts.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
pur_analysisage = totalbincounts
agedata3 = purchase_data.groupby('Total Count')
pur_analysisage['Average Purchase Price'] = round(agedata3['Price'].mean(),2)
pur_analysisage['Total Purchase Value'] = round(agedata3['Price'].sum(),2)
pur_analysisage['Avg Total Purchase per Person'] = round(pur_analysisage['Total Purchase Value']/pur_analysisage['Total Count'],2)

pur_analysisage['Avg Total Purchase per Person'] = pur_analysisage['Avg Total Purchase per Person'].map("${:.2f}".format)
pur_analysisage['Total Purchase Value'] = pur_analysisage['Total Purchase Value'].map("${:,.2f}".format)
pur_analysisage['Average Purchase Price'] = pur_analysisage['Average Purchase Price'].map("${:.2f}".format)

pur_analysisage = pur_analysisage.drop(columns=['Percentage of Players'])

pur_analysisage.sort_index()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#topspender

purchasecounts_sn = purchase_data.groupby('SN')['Purchase ID'].count()
avgprice_sn = round(purchase_data.groupby('SN')['Price'].mean(),2)
totprice_sn = round(purchase_data.groupby('SN')['Price'].sum(),2)

topspender = pd.DataFrame(purchasecounts_sn)
topspender = topspender.rename(columns = {'Purchase ID':'Purchase Count'})

topspender['Average Purchase Price'] = avgprice_sn
topspender['Total Purchase Value'] = totprice_sn
topspender = topspender.sort_values('Total Purchase Value', ascending = False)

topspender['Average Purchase Price'] = topspender['Average Purchase Price'].map("${:.2f}".format)
topspender['Total Purchase Value'] = topspender['Total Purchase Value'].map("${:.2f}".format)

topspender.head()
```
